# Single-phase flow and heat exchange

The aim  is to present the mathematical equations for single-phase fluid flow
in a porous medium coupled with heat exchange between the liquid and the surrounding
rock matrix.

## Working hypotheses

Let us suppose that the maximum reservoir depth is $z_{\max} \sim 5km$,
considering an approximate geothermal gradient of $25 ^\circ C \cdot km^{-1}$ the
maximum temperature reached at the bottom boundary is around $150^\circ C$. We
assume the range of temperature is $[10 ^\circ C, 150 ^\circ C]$.
Moreover, roughly speaking, the idrostatic pressure gradient
is $10 MPa \cdot km^{-1}$ obtaining a maximum of pressure at the bottom of the
reservoir of $50 MPa$.
<img src="./fig/water_phase_diagram_2.gif" width="45%">
At these condition, from the water-phase diagram reported in the previous Figure, 
the can assume that the water is only in the liquid state.
We assume also that the water does not react with the rock matrix and only an
heat exchange takes place, instantaneously with respect the typical time scale
of the simulation, between the water and the surrounding medium.

## Physical equations

Let us assume that the water is only present in the liquid phase and
the rock matrix is inert and at rest. Our objective is to compute the Darcy
velocity $\mathbf{u}$ $[m \cdot s^{-1}]$, the pressure $p$ $[Pa]$ and the temperature
$T$ $[^\circ C]$ such that: the Darcy problem is satisfied
\begin{gather}
    \begin{cases}
        \nabla \cdot \left( \rho_w \mathbf{u} \right) = -\phi \dfrac{\partial
        \rho_w}{\partial t} \\
        \mathbf{u} = - \dfrac{K}{\mu} \left( \nabla p - \rho_w g \mathbf{e}_3 \right)
    \end{cases},
\end{gather}
coupled with the thermal conduction and convection equation
\begin{gather} \label{eq:complete_thermal}
    c_e \dfrac{\partial T}{\partial t} + \rho_w
    c_w \mathbf{u} \cdot \nabla T = \nabla \cdot \Lambda_e \nabla T.
\end{gather}
In the previous equation the effective thermal capacity
$c_e$ $[J \cdot m^{-2} \cdot K^{-1}]$ is defined as the weighted averaged
between the water $c_w$ and solid $c_s$ specific thermal capacity, namely
\begin{gather}\label{eq:effective_capacty}
    c_e = \phi \rho_w c_w + \left(1-\phi\right) \rho_s c_s.
\end{gather}                                                                     
The data in the system are
* porosity $\phi$ $[\cdot]$ of the rock matrix;                                    
* density of rock matrix $\rho_s$ and water $\rho_w$ both $[kg \cdot m^{-3}]$;
* absolute permeability of the rock matrix $K$ $[m^2]$;
* dynamic viscosity $\mu$ $[Pa \cdot s]$ of water;
* modulus of gravity acceleration $g$ $[m \cdot s^{-2}]$ equal to $9.81 m \cdot s^{-2}$;
* specific heat capacity of water $c_w$ and solid $c_s$ both $[J \cdot kg^{-1} \cdot K^{-1}]$;
* effective thermal conductivity $\Lambda_e$ measured in $[W \cdot m^{-1} \cdot K^{-1}]$.

## Density

In this part we present empirical laws and relation among the density, water and solid rock, and the unknown of the flow.
The authors in [Somerton 1992](#Somerton1992) and Waples2004 [Waples 2004](#Waples2004) discuss an empirical law of
the density of the water related on the temperature, they propose the following
\begin{gather}
    \rho_w = \dfrac{\rho_w(T_0)}{1+\eta_w\left( T - T_0 \right)},
\end{gather}
where $\rho_w(T_0)$ is the density at a specific temperature value and $\eta$
$[\cdot]$ is the coefficient of the thermal expansion. For example at $T_0 =
10 ^\circ C$ we get $\rho_w(T_0) = 999.8349 kg \cdot m^{-3}$. The coefficient of
the thermal expansion for the water can be evaluated using
\begin{gather*}
    \eta_w = 0.0002115 + 1.32 \cdot 10^{-6} T + 1.09 \cdot 10^{-8} T^2.
\end{gather*}
The following Figure depicts the value of $\eta_w$ in terms of the temperature.
<img src="./fig/thermal_expansion.png" width="45%">
If the variation of the temperature is not severe, it is possible to approximate the previous
relation by the first order Taylor expansion obtaining
\begin{gather}
    \rho_w = \rho_w(T_0) \left[ 1 - \eta_w\left( T - T_0 \right) \right].
\end{gather}
This expression is considered, for example, in [Bear 1972](#Bear1972).
The next Figure presents a comparison between these two expressions of the water densisyt 
for our range of temperature.
<img src="./fig/density.png" width="45%">
We observe that the two expressions are in good agreement for values less than $100^\circ C$ and
a maximum error of $0.84\%$ at $150^\circ C$.

## Thermal conductivity

In this part we present a mathematical model to describe the effective thermal
conductivity $\Lambda_e$ introduced in the heat equation.
Let us assume that $\Lambda_e$ is an isotropic tensor such that $\Lambda_e =
\lambda_e I$, with $\lambda_e$ $[W \cdot m^{-1} \cdot K^{-1}]$ a scalar value.  Since it
represents a combination between the water $\lambda_w$ $[W \cdot m^{-1}  \cdot K^{-1}]$ and
solid $\lambda_s$ $[W \cdot m^{-1} \cdot K^{-1}]$ thermal conductivity,
following [Schon 2011](#Schon2011) we consider $\lambda_e$
defined through the weighted geometric average of both
\begin{gather*}
    \lambda_e = \lambda_w^\phi \lambda_s^{1-\phi}.
\end{gather*}
In the previous relation the solid is considered as a unique phase, for this
reason to accommodate heterogeneous rock matrix composed by $N_s$ of different
rocks, the solid head conductivity can be expressed using
\begin{gather*} %\label{eq:solidConductivity}
    \lambda_s = \dfrac{1}{V_s^t} \sum_{i=1}^{N_s} V_s^i \lambda^i_s
    \quad \text{with} \quad
    V_s^t = \sum_{i=1}^{N_s} V_s^i
\end{gather*}
where $V_s^i/V_s^t$ $[\cdot]$ is the fraction of a specific rock type $i$ in a
control volume and $\lambda^i_s$ $[W \cdot m^{-1} \cdot K^{-1}]$ is the head
conductivity of the rock $i$. For most rock types, the head conductivity
decreases with increasing temperature, and can be a function of the pressure $p$
and temperature $T$
\begin{gather*}
    \lambda_s^i = \lambda_s^i(T_0) \dfrac{1+\alpha_0^i p}{1 + c_0^i T},
\end{gather*}
where $\lambda_s^i(T_0)$ $[W \cdot m^{-1} \cdot K^{-1}]$ is a reference heat
conductivity at $T_0$, $c_0^i$ $[K^{-1}]$ and $\alpha_0^i$ $[Pa^{-1}]$ are
rock dependent parameters.  In [Griffiths 1992](#Griffiths1992) and in [Schon 2011](#Schon2011)
Subsubsection 9.2.2, the authors derived an empirical relationship for thermal
conductivity of water $\lambda_w$
\begin{gather*}
    \lambda_w = 0.56 + 0.002 T- 1.01 \cdot 10^{-5} T^2 +
    6.71 \cdot 10^{-9} T^3,
\end{gather*}
Moreover the influence of pressure on thermal conductivity of the water is relatively
small compared with the influence of temperature. The dependence of $\lambda_w$ from $T$ is depicted in 
the next Figure.

## Specific heat capacity
Let us consider the specific heat capacity of the rock matrix $c_s$ first. Since
the underground is heterogeneous, for a control volume composed by $N_s$ of
different rocks, we can define the specific heat capacity $c_s^i$ $[J \cdot
kg^{-1} \cdot K^{-1}]$ for each of them. Moreover, in our model, only one single value is considered we average
the specific heat capacity for all the rocks according to their mass $\rho_s^i
V_s^i$ in the reference volume
\begin{gather*}
    c_s = \dfrac{1}{m_s^t} \sum_{i=1}^{N_s} c_s^i \rho_s^i V_s^i
    \quad \text{with} \quad
    m_s^t = \sum_{i=1}^{N_s} \rho_s^i V_s^i.
\end{gather*}
We assume that the rock matrix is incompressible such that the specific heat
capacity at constant pressure is equal to the specific heat capacity at constant
volume. This assumption gives us the independence from the pressure, however
$c_s^i$ depends on the temperature in a linear way. Following
[Eppelbaum 2014](#Eppelbaum2014) Subsection 2.2 we obtain
\begin{gather*}
    c_s^i = c_s^i(T_0) + \zeta \left( T - T_0 \right),
\end{gather*}
where $T_0$ is the initial temperature and $\zeta$ $[J \cdot kg^{-1} \cdot
K^{-2}]$ is a rock dependent coefficient. For example are reported in the following Table.

| rock type |  $c_s$ at $10 ^\circ C$ | $c_s$ at $150 ^\circ C$ | $\zeta$               |
|-----------|---------------------|---------------------|-----------------------|
| sandstone | 823.82              | 834.06              | 8.9 1e-2              |
| shale     | 794.37              | 808.73              | 10.26 1e-2            |
| siltstone | 849.02              | 860.27              | 8.04 1e-2             |

An empirical relation for the water specific heat capacity $c_w$ is proposed in
[Waples 2004](#Waples2004), which is valid for values of temperature between
$20^\circ C - 290^\circ C$ and is
\begin{gather}
    c_w = \dfrac{4245 - 1.841 T}{\rho_w}
\end{gather}
where $\rho_w$ is computed using one of the previous model.
The previous equation can be extrapolated down to
$10 ^\circ C$ without significant error, and thus can be used for the temperature
interval we have assumed. The following Figure
shows the value of $c_w$ for different values of temperature.
<img src="./fig/specific_heat_capacity_water.png" width="45%">
As long as the pressure is high enough to keep the
water in a liquid phase as it is in our case, the specific heat capacity of
water under subsurface conditions can be estimated with good accuracy, without including pressure dependence.

## Dynamic viscosity

The dynamic viscosity of water is highly temperature dependent
\begin{gather}
    \mu = \mu_0 10^{ \delta / \left(T - 140 \right)}
\end{gather}
with $\mu_0 = 2.414 \cdot 10^{-5} Pa \cdot s$ and $\delta = 247.8$.
